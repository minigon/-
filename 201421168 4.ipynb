```python
%matplotlib inline
from numpy import *
from pylab import *
#word dispersion, 한 단어가 얼마나 많이 나왔는가
text4
l=len(text4)
for word in range(0,l):
    a=text4[word].lower()
    if a == 'citizens':
        plot(word,1,'b.')
    if a == 'world':
        plot(word,2,'r.')
    if a == 'war':
        plot(word,3,'y.')
    if a == 'peace':
        plot(word,4,'g.')
    if a == 'freedom':
        plot(word,5,'b.')
    if a == 'economy':
        plot(word,6,'r.')
    if a == 'america':
        plot(word,7,'y.')
axis([0,l,0,8])
xlabel('word number')
show()
```

```python
from nltk.corpus import inaugural
inaugural.fileids()
#가장 길게 말한 사람
n=1
for i in inaugural.fileids():
    word=len(inaugural.words(i))
    plot(n,word,'o')
    n=n+1
axis([0,60,0,10000])
ylabel('length of speech')
show()
```

```python
#가장 많은 단어를 쓴 사람
n=0
for i in inaugural.fileids():
    n=n+1
    v=set(w.lower() for w in inaugural.words(i))
    nv=len(v)
    plot(n,nv,'o')
show()
```

```python
n=0
for i in inaugural.fileids():
    n = n+1
    #character(알파벳)
    c = len(inaugural.raw(i))
    w = len(inaugural.words(i))
    #sentence(문장)
    s = len(inaugural.sents(i))
    nv=set(w.lower() for w in inaugural.words(i))
    v=len(nv)
    
    #char/word (한 단어에 평균적으로 몇 개의 알파벳이 들어가는가)
    ave_c=c/w
    #word/sentence (한 문장에 평균적으로 몇 개의 단어가 들어가는가)
    ave_w=w/s
    #freq/word(voca) (한 단어가 평균적으로 몇 번 반복되는가)
    freq_w=w/v
    plot(n,ave_c,'mo')
    plot(n,ave_w,'bo')
    plot(n,freq_w,'ro')
show()
```

```python
#모든 연설문에 공통으로 나온 단어
v=set(w.lower() for w in inaugural.words(inaugural.fileids()[0]))
for i in inaugural.fileids():
    vc = set(w.lower() for w in inaugural.words(i))
    v = set.intersection(v,vc)
print(v)
```

```python
#각 단어를 각 대통령마다 몇 번씩 썼는지
n=0
for i in inaugural.fileids():
    n=n+1
    w=inaugural.raw(i).lower()
    war=w.count('war')
    america=w.count('america')
    economy=w.count('economy')
    world=w.count('world')
    
    plot(n,war,'ko')
    plot(n,america,'bo')
    plot(n,economy,'ro')
    plot(n,world,'go')
xlabel('t')
legend(['war','america','economy','world'])
show()
```

```python
from pylab import *
from numpy import *
from nltk.corpus import PlaintextCorpusReader
my_dir="/Users/gktjd/exercise"
my_list=PlaintextCorpusReader(my_dir,'.*\.txt')

files = my_list.fileids()
n = len(files)

ave_char=[]
ave_word=[]
freq_vocab=[]

for i in my_list.fileids():
    c = len(my_list.raw(i))
    w = len(my_list.words(i))
    s = len(my_list.sents(i))
    v = len(set(w.lower() for w in my_list.words(i)))
    
    green = c/w #char/word
    blue = w/s
    red = w/v
    ave_char.append(green)
    ave_word.append(blue)
    freq_vocab.append(red)
    
plot(range(0,n),ave_char,'-og',label='char/word')
plot(range(0,n),ave_word,'-ob',label='word/sent')
plot(range(0,n),freq_vocab,'-or',label='word/voca')
xticks(arange(n),(files))
legend()
show()
```

```python
n=len(files)
a_vocab = set(w for w in my_list.words('a.txt'))
text1 = my_list.open('a.txt').read()
vocab_list = list(a_vocab)
t = len(a_vocab)
c_vocab = zeros((t,2),dtype = object)
for i in range(t):
    voc = vocab_list[i]
    ct = text1.count(voc)
    c_vocab[i,0] = ct
    c_vocab[i,1] = voc
def getkey(item):
    return item[0]
sorted(c_vocab,key = getkey, reverse=True)
```

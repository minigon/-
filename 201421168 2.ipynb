```python
from numpy import *
from pylab import *
import pandas as pd
from sklearn import svm
import seaborn as sns
%matplotlib inline
```

```python
example1=pd.read_csv('test2.csv')
example1.head()
```

```python
D=example1[['x1','x2']].as_matrix()
class_label=where(example1['class']=='neg',0,1)

model = svm.SVC(kernel='linear',C=1)
model.fit(D, class_label)
w = model.coef_[0]
a = -w[0]/w[1]
x = linspace(-5,6)
y = a*x - (model.intercept_[0])/w[1]
b = model.support_vectors_[0]
n_margin = a*x + (b[1]-a*b[0])
b = model.support_vectors_[-1]
p_margin = a*x + (b[1]-a*b[0])

plot(x,y,'r-')
plot(x,n_margin,'k--')
plot(x,p_margin,'k--')
plot(example1['x1'],example1['x2'],'or')
sns.lmplot('x1','x2',data=example1, hue='class', palette='Set1', fit_reg=False, scatter_kws={"s":70})
```

```python
D=example1[['x1','x2']].as_matrix()
class_label=where(example1['class']=='neg',0,1)

model = svm.SVC(kernel='rbf',C=1,gamma=1)
model.fit(D, class_label)
xx,yy=meshgrid(linspace(-6,6,200),linspace(-6,6,200))
z=model.decision_function(np.c_[xx.ravel(),yy.ravel()])
z=z.reshape(xx.shape)
pcolormesh(xx,yy,-z)
plot(D[0:10,0],D[0:10,1],'ob')
plot(D[10:,0],D[10:,1],'or')
```

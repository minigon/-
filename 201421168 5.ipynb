```python
import urllib.request
wiki = "https://en.wikipedia.org/wiki/South_Korea"
page = urllib.request.urlopen(wiki)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page,"lxml")
print(soup.prettify())
```

```python
#사이트 제목 출력
print(soup.title)
print(soup.title.string)
```

```python
print(soup.a)
```

```python
#a가 들어간 거 전부 찾기
soup.find_all("a")
```

```python
#href가 들어가는 거 전부 출력
url_list = []
all_url = soup.find_all("a")
for link in all_url:
    href_link = link.get('href')
    print(href_link)
    url_list.append(href_link)
```

```python
#href 들어간 거
url_test = []
#https 들어간 거
url_test2 = []
all_url = soup.find_all("a")
for link in all_url:
    test = link.get('href')
    url_test.append(test)
    if "http" in str(test):
        url_test2.append(test)
        print(test)
```

```python
from pandas import *
wiki = "https://en.wikipedia.org/wiki/South_Korea"
page = urllib.request.urlopen(wiki)
soup = BeautifulSoup(page,"lxml")
#href 갯수
url_href = []
#http 갯수
url_http = []
#http 중 kr은 1 아니면 0
url_kr = []
url_kr1 = []
all_url = soup.find_all("a")
for link in all_url:
    test = link.get('href')
    url_href.append(test)
    if "http" in str(test):
        url_http.append(test)
        if ".kr" in str(test):
            url_kr.append(1)
            url_kr1.append(1)
        else:
            url_kr.append(0)
url_table = DataFrame(url_http,columns=['External lin'])
url_table['.kr'] = url_kr
print(url_table)
test_table = url_table.describe()
print(len(url_kr1))
print(test_table)
```
